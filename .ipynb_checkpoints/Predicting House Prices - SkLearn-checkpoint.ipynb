```python
import pandas as pd
from sklearn import linear_model
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
houses = pd.read_csv('data/home-data.csv')
houses
```

# Using only one feature

```python
features = houses['sqft_living']
target = houses['price']
plt.plot(features, target, ".")
```

```python
train_features = np.array(features[0:20000].as_matrix())
test_features = np.array(features[20000:].as_matrix())
train_targets = np.array(target[0:20000].as_matrix())
test_targets = np.array(target[20000:].as_matrix())
```

```python
regression = linear_model.LinearRegression()
```

```python
reshaped_train_data = train_features.reshape(-1, 1)
print(train_features)
print(reshaped_train_data)

regression.fit(reshaped_train_data, train_targets)
```

```python
# The coefficients
print('Coefficients: \n', regression.coef_)
```

```python
# The mean square error
print("Residual sum of squares: %.2f"
      % np.mean((regression.predict(test_features.reshape(-1,1)) - test_targets) ** 2))
```

```python
plt.scatter(test_features, test_targets,  color='black')
plt.plot(test_features, regression.predict(test_features.reshape(-1,1)), color='blue',
         linewidth=3)
```

# Using more features

```python
more_features = houses[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode']]
more_features
```

```python
train_features = np.array(more_features[0:20000].as_matrix())
test_features = np.array(more_features[20000:].as_matrix())
train_targets = np.array(target[0:20000].as_matrix())
test_targets = np.array(target[20000:].as_matrix())
```

```python
regression = linear_model.LinearRegression()
reshaped_train_data = train_features.reshape(-1, 6)
print(train_features)
print(reshaped_train_data)

regression.fit(reshaped_train_data, train_targets)
```

```python
# The coefficients
print('Coefficients: \n', regression.coef_)
# The mean square error
print("Residual sum of squares: %.2f"
      % np.mean((regression.predict(test_features.reshape(-1,6)) - test_targets) ** 2))
```

```python
house1 = houses[houses['id'] == 5309101200]
print("Real price U$%f" % house1['price'].values[0])
print("Predict price U$%.2f" %regression.predict(house1[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode']])[0])
```
